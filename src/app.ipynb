# 5 SPACE 

```python
def remove(lst):return [item.replace(" ", "") for item in lst] if isinstance(lst, list) else lst

newtable[["genres", "cast", "keywords"]] = newtable[["genres", "cast", "keywords"]].applymap(remove)
newtable["crew"] = newtable["crew"].str.replace(" ", "")

newtable.head()
```

# 6 CHANGE JSON: -LIST GO STRING

```python
# Función para convertir una lista a una cadena separada por comas
def list_to_string(lst):
    # Verificar si el valor es una lista
    if isinstance(lst, list):
        # Si es una lista, convertirla en una cadena separada por comas
        return ",".join(lst)
    else:
        # Si no es una lista, mantener el valor sin cambios
        return lst

# Aplicar la función a las columnas "genres", "keywords", "cast" y "overview" de la tabla "newtable"
newtable["genres"] = newtable["genres"].apply(list_to_string)
newtable["keywords"] = newtable["keywords"].apply(list_to_string)
newtable["cast"] = newtable["cast"].apply(list_to_string)
newtable["overview"] = newtable["overview"].apply(list_to_string)
```

```python
newtable.head()
```

# 7 NEW COLUMN: TAGS

```python
newtable["tags"] = newtable["overview"] + " " + newtable["genres"] + " " + newtable["keywords"] + " " + newtable["cast"] + " " + newtable["crew"]

newtable.iloc[0].tags
```

# 8 MACHINE LEARNING KNN

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity

vector = CountVectorizer(max_features=5000, stop_words='english')
vectors = vector.fit_transform(newtable['tags'].fillna("")).toarray()
similarity_matrix = cosine_similarity(vectors)
def recommend(movie):
    movie_low = movie.lower()
    movie_index = newtable[newtable["title"].str.lower() == movie_low].index[0]
    movie_distances = similarity_matrix[movie_index]
    similar_movies_indices = movie_distances.argsort()[::-1][1:6]
    
    print(f"Películas similares a '{movie}':")
    for index in similar_movies_indices:
        print(newtable.iloc[index].title)

x = input("Introduce una película: ")
recommend(x)
```
